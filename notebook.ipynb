## Visualization & Explanatory Analysis

### Pertanyaan 1: Apakah pengguna sepeda lebih cenderung keluar saat cuaca cerah atau saat cuaca buruk?

```python
# Menyalin data dari hour_df
weather_plot = hour_df[['weather_condition', 'total_user']]

# Mengganti data weather_condition dengan kondisi dominannya
weather_plot.replace({
    'weather_condition': {
        1: 'Clear',
        2: 'Mist',
        3: 'Light Snow, Light Rain',
        4: 'Heavy Rain, Snow, Fog'
    }
}, inplace=True)

# Agregat total pengguna berdasarkan weather_condition
weather_plot = weather_plot.groupby('weather_condition')['total_user'].sum().reset_index()

# Urutkan tabel berdasarkan total pengguna dengan urutan menurun
weather_plot = weather_plot.sort_values('total_user', ascending=False)

# Definisikan palet warna
color_list = ['#1565c0', '#bbdefb', '#bbdefb', '#bbdefb']

# Buat barplot
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 5))
sns.barplot(data=weather_plot, x='total_user', y='weather_condition', palette=color_list)

plt.title("Total Bike Users in Different Weather Conditions")
plt.ylabel(None)
plt.xlabel("Total Users")

# Tambahkan data label
for p in ax.patches:
    width = p.get_width()
    plt.text(width, p.get_y() + p.get_height() / 2, f"{width:,.0f}", ha='left', va='center')

plt.tight_layout()
plt.show()
```

### Pertanyaan 2: Musim apa yang menjadi favorit pengguna untuk bersepeda?

```python
color_list = ['#bbdefb', '#bbdefb', '#1565c0', '#bbdefb']

# Buat barplot
sns.barplot(data=hour_df.groupby('season')['total_user'].sum().reset_index(), 
            x='season', y='total_user',
            palette=color_list)

plt.title("Most Favourite Season to Ride a Bike")
plt.xlabel('Season')
plt.ylabel('Total Users')

plt.tight_layout()
plt.show()
```

### Pertanyaan 3: Bagaimana performa penjualan perusahaan selama beberapa tahun terakhir?

```python
# Menyalin data dari day_df
monthly_counts = day_df.groupby('datetime')['total_user'].max().reset_index()

# Membuat figure dengan ukuran (24, 5)
plt.figure(figsize=(24, 5))

# Membuat scatter plot untuk jumlah pelanggan maksimum per bulan, dengan warna biru dan marker 'o'
sns.scatterplot(x='datetime', y='total_user', data=monthly_counts, color="red", s=10, marker='o')

# Membuat line plot untuk jumlah pelanggan maksimum per bulan
sns.lineplot(x='datetime', y='total_user', data=monthly_counts)

# Penamaan untuk sumbu x
plt.xlabel('Bulan')

# Penamaan untuk sumbu y
plt.ylabel('Jumlah')

# Judul plot
plt.title('Grafik Jumlah Pelanggan per Bulan pada Tahun 2012')

# Menampilkan plot
plt.show()
```
